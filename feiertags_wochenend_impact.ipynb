# Verzögerung der Müllabholung durch Samstage, Sonntage, Feiertage

```python
import pandas as pd

df_feiertage = pd.read_csv("data/feiertage_berlin_2019_2023.csv")
df_wochentage = pd.read_csv("data/wochentage_numerisch.csv")
```

```python
df_feiertage.head()
```

```python
df_wochentage.head()
```

```python
# Merge der beiden DataFrames
df = pd.merge(df_wochentage, df_feiertage, on='Datum', how='left')  # Feiertage werden in den Wochentag-Datensatz integriert
```
